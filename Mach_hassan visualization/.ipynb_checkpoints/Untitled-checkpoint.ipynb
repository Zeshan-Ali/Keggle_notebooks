```python
# Plotting the data
plt.figure(figsize=(12, 6))
sns.barplot(data=df, x='Area /Machine', y='Tonnage', palette='viridis')
plt.title('Tonnage Distribution by Area / Machine')
plt.xlabel('Area / Machine')
plt.ylabel('Tonnage')
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability if necessary
plt.tight_layout()  # Adjust layout for better fit
plt.show()
```

```python
# Get unique sections
sections = df['Section'].unique()

# Plotting each section separately
for section in sections:
    # Filter data for the current section
    section_df = df[df['Section'] == section]
    
    # Plotting the data
    plt.figure(figsize=(12, 6))
    sns.barplot(data=section_df, x='Area /Machine', y='Tonnage', palette='viridis')
    plt.title(f'Tonnage Distribution for Section: {section}')
    plt.xlabel('Area / Machine')
    plt.ylabel('Tonnage')
    plt.xticks(rotation=45)  # Rotate x-axis labels for better readability if necessary
    plt.tight_layout()  # Adjust layout for better fit
    plt.show()
```

```python
# Get unique sections and brands
sections = df['Section'].unique()
brands = df['Brand'].unique()

# Define brands with fewer data points
brands_together = ['NSK Japan', 'Insulated Bearing German SKF','NTN']  # Update with other brands as needed

# Plotting each section and brand combination separately
for section in sections:
    for brand in brands:
        # Filter data for the current section and brand
        section_brand_df = df[(df['Section'] == section) & (df['Brand'] == brand)]
        
        if not section_brand_df.empty:
            # Determine whether to plot together or separately based on brand count
            if brand in brands_together or section_brand_df.shape[0] <= 1:
                # Plot together in one graph
                plt.figure(figsize=(12, 6))
                ax = sns.barplot(data=section_brand_df, x='Area /Machine', y='Tonnage', palette='viridis', errorbar=None)
                plt.title(f'Tonnage Distribution for Section: {section}, Brand: {brand}')
                plt.xlabel('Area / Machine')
                plt.ylabel('Tonnage')
                plt.xticks(rotation=90)  # Rotate x-axis labels for better readability if necessary
                
                # Annotate each bar with its tonnage value
                for p in ax.patches:
                    ax.annotate(f'{p.get_height():.2f}', 
                                (p.get_x() + p.get_width() / 2., p.get_height()), 
                                ha='center', va='center', xytext=(0, 10), 
                                textcoords='offset points', fontsize=10)
                
                plt.tight_layout()  # Adjust layout for better fit
                plt.show()
            else:
                # Plot separately in individual graphs
                plt.figure(figsize=(12, 6))
                ax = sns.barplot(data=section_brand_df, x='Area /Machine', y='Tonnage', palette='viridis', errorbar=None)
                plt.title(f'Tonnage Distribution for Section: {section}, Brand: {brand}')
                plt.xlabel('Area / Machine')
                plt.ylabel('Tonnage')
                plt.xticks(rotation=90)  # Rotate x-axis labels for better readability if necessary
                
                # Annotate each bar with its tonnage value
                for p in ax.patches:
                    ax.annotate(f'{p.get_height():.2f}', 
                                (p.get_x() + p.get_width() / 2., p.get_height()), 
                                ha='center', va='center', xytext=(0, 10), 
                                textcoords='offset points', fontsize=10)
                
                plt.tight_layout()  # Adjust layout for better fit
                plt.show()
```
