```python
import tensorflow as tf
import numpy as np
import random

# Set random seed for TensorFlow
tf.random.set_seed(42)

# Set random seed for NumPy
np.random.seed(42)

# Set random seed for Python
random.seed(42)

import os, sys, datetime
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
import tensorflow as tf
from tensorflow.keras import *
import keras
import h5py

from random import shuffle


from IPython.display import display
from IPython.display import Image as _Imgdis
from sklearn.model_selection import train_test_split
from keras.preprocessing.image import ImageDataGenerator, load_img, img_to_array
```

```python
effusion_train_test_dataset_dir = '/content/drive/MyDrive/dataset/train_test/effusion'
normal_train_test_dataset_dir = '/content/drive/MyDrive/dataset/train_test/normal'
```

```python
all_effusion_images = [f for f in os.listdir(effusion_train_test_dataset_dir) if os.path.isfile(os.path.join(effusion_train_test_dataset_dir, f))]
all_normal_images = [f for f in os.listdir(normal_train_test_dataset_dir) if os.path.isfile(os.path.join(normal_train_test_dataset_dir, f))]

print("Number of effusion samples: "+str(len(all_effusion_images)))
print("Number of normal samples: "+str(len(all_normal_images)))
```

```python
for i in range(0,1):
  img = mpimg.imread(effusion_train_test_dataset_dir + "/" + all_effusion_images[i])
  imgplot = plt.imshow(img)
  plt.show()
```

```python
for i in range(0,1):
  img = mpimg.imread(normal_train_test_dataset_dir + "/" + all_normal_images[i])
  imgplot = plt.imshow(img)
  plt.show()
```

```python
# combining normal and effusion samples

all_samples = []
effusion_samples = []
normal_samples = []

for sample in all_effusion_images:
  effusion_samples.append(effusion_train_test_dataset_dir + "/" + sample)

for sample in all_normal_images:
  normal_samples.append(normal_train_test_dataset_dir + "/" + sample)

all_samples = effusion_samples + normal_samples
shuffle(all_samples)
```

```python
print(effusion_samples[0])
print(normal_samples[0])
print("Total samples: " + str(len(all_samples)))
```

```python
# splitting images into features and labels
# loading all images into an ndarray

features_dirs = all_samples
y = []

channels = 1
im_height = 64
im_width = 64

image_array = np.ndarray(shape = (len(all_samples),
                                  im_height,
                                  im_width,
                                  channels),
                         dtype = np.float32)
i = 0

for sample in features_dirs:

    img = load_img(sample, grayscale=True)
    img.thumbnail((im_width, im_height))
    x = img_to_array(img)
    x = x.reshape((64, 64, 1))
    image_array[i] = x
    i+=1

    if 'normal' in sample:
      y.append(0)
    else:
      y.append(1)

y = np.array((y))
```

```python
# saving loaded dataset as a

with h5py.File('X.h5', 'w') as X_hdf5:
  X_hdf5.create_dataset('X', data = image_array)

with h5py.File('y.h5', 'w') as y_hdf5:
  y_hdf5.create_dataset('y', data = y)
```

```python
with h5py.File('X.h5', 'r') as hf:
    image_array_h5 = hf['X'][:]

with h5py.File('y.h5', 'r') as hf:
    y_h5 = hf['y'][:]

print("Features shape: " + str(image_array_h5.shape))
print("Labels shape: " + str(y_h5.shape))
```

```python
X_train, X_test, y_train, y_test = train_test_split(image_array, y, test_size = 0.2, random_state = 0)
```

```python
num_classes = 2
y_train = tf.keras.utils.to_categorical(y_train, num_classes)
y_test = tf.keras.utils.to_categorical(y_test, num_classes)
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.callbacks import LearningRateScheduler, EarlyStopping
from tensorflow.keras.layers import BatchNormalization, Dropout, Conv2D, MaxPooling2D, Flatten, Dense
from tensorflow.keras.models import Sequential
from tensorflow.keras import regularizers
import numpy as np

# Define data augmentation parameters
datagen = ImageDataGenerator(
    rotation_range=60,
    width_shift_range=0.05,
    height_shift_range=0.05,
    shear_range=0.05,
    zoom_range=0.05,
    horizontal_flip=True,
    vertical_flip=True,
    brightness_range=[0.5, 1.5],  # Add brightness adjustment
    channel_shift_range=50.0,
    fill_mode='nearest'
)

# Define learning rate schedule
def lr_schedule(epoch):
    lr = 0.001
    if epoch > 15:
        lr *= 0.5
    elif epoch > 10:
        lr *= 0.75
    return lr


model = Sequential([
    Conv2D(64, kernel_size=(3, 3), activation='relu', input_shape=(64, 64, 1)),
    BatchNormalization(),
    Conv2D(128, (3, 3), activation='relu'),
    BatchNormalization(),
    MaxPooling2D(pool_size=(2, 2)),
    Dropout(0.5),

    Conv2D(256, (3, 3), activation='relu'),
    BatchNormalization(),
    Conv2D(512, (3, 3), activation='relu'),
    BatchNormalization(),
    MaxPooling2D(pool_size=(2, 2)),
    Dropout(0.5),

    Flatten(),
    Dense(512, activation='relu'),
    BatchNormalization(),
    Dropout(0.5),
    Dense(256, activation='relu'),
    BatchNormalization(),
    Dropout(0.5),
    Dense(num_classes, activation='softmax')
])

 #Compile the model
model.compile(loss='categorical_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])

# Data augmentation and training
history = model.fit(datagen.flow(X_train, y_train, batch_size=32),
                    steps_per_epoch=len(X_train) / 32,
                    epochs=25,
                    verbose=1,
                    validation_data=(X_test, y_test),
                    callbacks=[LearningRateScheduler(lr_schedule), EarlyStopping(patience=10)])

# Evaluate the model
score = model.evaluate(X_test, y_test, verbose=0)
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```

```python
def plot_figures(figures, n_rows, n_cols):

    fig, axes = plt.subplots(ncols = n_cols,
                                 nrows = n_rows)
    for ind, title in enumerate(figures):
        axes.ravel()[ind].imshow(figures[title],
                                     cmap=plt.gray())
        axes.ravel()[ind].set_title(title)
        axes.ravel()[ind].set_axis_off()
    plt.tight_layout()
```

```python

def img_to_numpy_array(img_object):
  img = load_img(img_object, grayscale=True)
  img.thumbnail((im_width, im_height))
  x = img_to_array(img)
  x = x.reshape((1, 64, 64, 1))
  return x
```

```python
test_effusion_dir = '/content/drive/MyDrive/dataset/validation/effusion'
test_normal_dir = '/content/drive/MyDrive/dataset/validation/normal'

total_effusion_test_images = [f for f in os.listdir(test_effusion_dir) if os.path.isfile(os.path.join(test_effusion_dir, f))]
total_normal_test_images = [f for f in os.listdir(test_normal_dir) if os.path.isfile(os.path.join(test_normal_dir, f))]

effusion_figures_list = []
normal_figures_list = []
prediction_effusion = []
prediction_normal = []

for i in range(len(total_effusion_test_images)):
  effusion_img_path = test_effusion_dir + '/' + total_effusion_test_images[i]
  normal_img_path = test_normal_dir + '/' + total_normal_test_images[i]
  img1 = mpimg.imread(effusion_img_path)
  img2 = mpimg.imread(normal_img_path)
  effusion_array = img_to_numpy_array(effusion_img_path)
  normal_array = img_to_numpy_array(normal_img_path)
  prediction_effusion.append(1 if model.predict(effusion_array)[0][0] > 0.5 else 0)
  prediction_normal.append(1 if model.predict(normal_array)[0][0] > 0.5 else 0)
  effusion_figures_list.append(img1)
  normal_figures_list.append(img2)

number_of_im = 4
effusion_figures = {'e '+str(i)+'predicted '+str(prediction_effusion[i]): effusion_figures_list[i] for i in range(number_of_im)}
normal_figures = {'n '+str(i)+'predicted '+str(prediction_normal[i]): normal_figures_list[i] for i in range(number_of_im)}

plot_figures(effusion_figures, 1, 4)
plot_figures(normal_figures, 1, 4)
```

```python
# applying transfer learning o nthe data
import numpy as np

# Convert grayscale images to RGB
X_train_rgb = np.repeat(X_train, 3, axis=-1)  # Repeat the single channel 3 times
X_test_rgb = np.repeat(X_test, 3, axis=-1)

# Now, you can use X_train_rgb and X_test_rgb for training and testing
```

```python
#applying transfer learning model resnet50
from tensorflow.keras.applications import ResNet50
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Flatten, Dense, Dropout
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# Data augmentation
train_datagen = ImageDataGenerator(
    rotation_range=40,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    vertical_flip=True,
    brightness_range=[0.5, 1.5],
    fill_mode='nearest'
)

# Load the pre-trained ResNet50 model without the top (fully connected) layers
base_model = ResNet50(weights='imagenet', include_top=False, input_shape=(64, 64, 3))

# Freeze the weights of the pre-trained layers
for layer in base_model.layers:
    layer.trainable = False

# Add custom top layers for classification
x = Flatten()(base_model.output)
x = Dense(256, activation='relu')(x)
x = Dropout(0.8)(x)  # Add dropout for regularization
x = Dense(num_classes, activation='softmax')(x)

# Create the final model
model = Model(inputs=base_model.input, outputs=x)

# Compile the model
model.compile(optimizer=Adam(learning_rate=0.0001), loss='categorical_crossentropy', metrics=['accuracy'])

# Train the model with data augmentation
history = model.fit(train_datagen.flow(X_train_rgb, y_train, batch_size=32),
                    steps_per_epoch=len(X_train_rgb) / 32,
                    epochs=10,
                    validation_data=(X_test_rgb, y_test))

# Print final value of accuracy
final_training_accuracy = history.history['accuracy'][-1]
final_validation_accuracy = history.history['val_accuracy'][-1]
print("Final Training Accuracy:", final_training_accuracy)
print("Final Validation Accuracy:", final_validation_accuracy)

# Check if overfitting
if final_training_accuracy > final_validation_accuracy:
    print("The model might be overfitting.")
else:
    print("The model is not overfitting.")
```

```python
# Evaluate the model on the test dataset
test_loss, test_accuracy = model.evaluate(X_test_rgb, y_test)
print("Test Loss:", test_loss)
print("Test Accuracy:", test_accuracy)
```

```python
# Make predictions on the test dataset
predictions = model.predict(X_test_rgb)

# Convert predictions to class labels
predicted_classes = np.argmax(predictions, axis=1)

# Display some sample predictions
for i in range(10):  # Displaying predictions for the first 10 samples
    print("Sample", i+1)
    print("Predicted Class:", predicted_classes[i])
    print("True Class:", np.argmax(y_test[i]))
    print()  # Adding a newline for clarity
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Convert true labels to 1D array (if not already)
true_labels = np.argmax(y_test, axis=1)

# Calculate accuracy
accuracy = accuracy_score(true_labels, predicted_classes)
print("Accuracy:", accuracy)

# Calculate precision
precision = precision_score(true_labels, predicted_classes)
print("Precision:", precision)

# Calculate recall
recall = recall_score(true_labels, predicted_classes)
print("Recall:", recall)

# Calculate F1-score
f1 = f1_score(true_labels, predicted_classes)
print("F1-score:", f1)
```
